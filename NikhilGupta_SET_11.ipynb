```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def main():
    img = cv2.imread("/content/restro1.jpg")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    k = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), np.float32)
    print(k)
    print(type(k))


#Output will be same as  k = identy kernal
    output = cv2.filter2D(img, -1, k)

    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Original Image')

    plt.subplot(1, 2, 2)
    plt.imshow(output)
    plt.title('sharp Masking Image')
    
    plt.show()


main()
```
